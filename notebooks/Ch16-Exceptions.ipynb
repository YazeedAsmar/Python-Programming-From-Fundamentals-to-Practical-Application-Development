## category of bugs
- syntax errors
- logical/semantic errors
- runtime errors/exceptions

Here are the explanations for each category of bugs:

### Syntax Errors
These are errors where the rules of the programming language are violated. Python will immediately detect these errors before the code even runs, and it won't be able to execute the program. Examples include forgetting a colon, misspelling a keyword, or an unmatched parenthesis. The Python interpreter will usually point to the line where the error occurred and give a `SyntaxError`.

### Logical/Semantic Errors
These are errors where the program runs without crashing, but it produces an incorrect or unexpected result. The program's syntax is correct, but its logic is flawed, meaning it doesn't do what the programmer intended. These are often the hardest bugs to find because Python doesn't complain; it just gives you the wrong answer. Debugging these requires careful examination of the code's behavior and expected output.

### Runtime Errors/Exceptions
These errors occur while the program is running, even if the syntax is correct and the initial logic seems fine. Something unexpected happens during execution that prevents the program from continuing normally. When a runtime error occurs, Python creates and "throws" an exception object. If this exception is not handled (caught), the program will halt, and Python will print a traceback with the exception name and message (e.g., `ZeroDivisionError`, `IndexError`, `TypeError`). The notebook examples `print(55/0)` (ZeroDivisionError) and `print(alist[0])` (IndexError) are perfect illustrations of runtime errors.

## catching exceptions
- use try and except blocks
- try statement has several separate clauses/parts
<pre>
    try:
        # statement(s) thay may potentially throw(s) exception
    except ExceptionName1:
        # catch specific exception ExceptionName1
        # statement(s) to handle the exception
    [except ExceptionName2 as err:
        # statements
    ]
    [except:
        # catch any error not caught by previous except blocks
    ]
    [else:
        # follows all except clause
        # executes if try clause does NOT raise an exception
    ]
    [finally:
        # clean-up actions that must be executed under all circumstances;
        # exectued on the way out when try block is left via a break, continue, or return statement
    ]
</pre>
- <em>[ ] optional </em>

### example 1

```python

try:
    # invalid type
    x = int(input("Enter dividend: "))
    # zero division
    y = int(input("Enter divisor: "))
    # lst = [0, 5, 7]
    quotient = x/y
    remainder = x%y
    # index error
    # print(lst[5])
except ZeroDivisionError as ex:
    print('Exception occured:', ex)
except ValueError as ex:
    print(ex)
# except IndexError as ex:
#     print(ex)
except Exception as ex:
    print('Some exception occured...', ex)
else:
    print("quotient=", quotient)
    print("remainder=", remainder)
finally:
    print("executing finally clause")
```

This code cell demonstrates Python's exception handling mechanism using `try`, `except`, `else`, and `finally` blocks.

*   **`try` block**: This is where you place the code that might raise an exception. In this case, it attempts to get two integer inputs from the user (`x` and `y`) and then performs division (`x/y`) and modulo (`x%y`) operations. If `y` is zero, a `ZeroDivisionError` will occur. If the input for `x` or `y` is not a valid integer, a `ValueError` will occur.

*   **`except ZeroDivisionError as ex:`**: This block catches a specific type of exception: `ZeroDivisionError`. If the user enters `0` for the divisor (`y`), this block will execute. It prints a message indicating the exception and its arguments.

*   **`except ValueError as ex:`**: This block catches `ValueError`. If the user enters non-integer input for either the dividend or divisor, this block will execute, printing the exception message.

*   **`except Exception as ex1:`**: This is a more general `except` block. It catches any other exception that is not specifically `ZeroDivisionError` or `ValueError`. It's good practice to catch `Exception` (or a more specific base class) to handle unforeseen errors. This block will print a generic message and the specific exception that occurred.


*   **`else` block**: This block executes *only if* the code in the `try` block completes successfully without raising any exceptions. Here, if no errors occur during input or calculation, it will print the `quotient` and `remainder`.

*   **`finally` block**: This block *always* executes, regardless of whether an exception occurred or not, or if it was handled. It's typically used for cleanup actions, such as closing files or releasing resources. In this example, it simply prints "executing finally clause".

## user-defined exceptions
- one can define their own exceptions and raise them as needed
- should typically derive from the Exception class, either directly or indirectly

### example 1

```python
class InputError(Exception):
    """
    Exception raised for errors in the input.

    Attributes:
    expression -- input expression in which the error occured
    message -- explaination of the error
    """
    def __init__(self, expression, message):
        self.expression = expression
        self.message = message
```

```python
help(InputError)
```

```python
def getInteger():
    x = input('Enter an integer number: ')
    if not x.isdigit():
        raise InputError(x, 'That is not an integer!')
    return int(x)
```

```python
x = getInteger()
print(x)
```

## catch user-defined exception

```python
try:
    x = getInteger() #may throw InputError
except InputError as ie:
    print('Exception:', ie)
    # can throw ie again
else:
    print('{}^2 = {}'.format(x, x**2))
```
